2)  We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.

   The response variable will be the total duration of hours played along with the number of sessions that each user did. The explanatory variable will be the player type (A catagorical variable describing the players experience level).

To work with the data I will first arrange my response variable in decending order to find which player type is has played the most and had the most sessions. I will then group my data by the player type that is most likely to have gathered the most data

```python
import pandas as pd
import altair as alt
players = pd.read_csv("data/players.csv")  #This is Tidy
sessions = pd.read_csv("data/sessions.csv")

sessions_tidy = pd.melt(sessions, id_vars=['hashedEmail'], 
                        value_vars=['start_time', 'end_time', 'original_start_time', 'original_end_time'],
                  var_name='time_type', value_name='value')
sessions_tidy

players
```

```python
players_filter = players[['experience', 'played_hours']]


grouped = players_filter.groupby('experience').mean().reset_index()


bar_plot = alt.Chart(grouped).mark_bar().encode(
    x=alt.X('experience').title('Experience'),
    y=alt.Y('played_hours').title('Average Played Hours')
)
    


bar_plot
```

This plot shows that players with a regular experience seems to produce on averageage the most amount of hours played and veteran seems to be the least hours played
